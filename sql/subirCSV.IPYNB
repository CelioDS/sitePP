# CRIAR TABELA

```python
CREATE TABLE db_projetos.pdu (
    ID INT AUTO_INCREMENT PRIMARY KEY,

    DT         VARCHAR(10)  NOT NULL,   -- 01/01/2026
    AAAA       INT          NOT NULL,   -- 2026
    MMM        VARCHAR(10)  NOT NULL,   -- JAN
    MM         INT          NOT NULL,   -- 1
    DD         INT          NOT NULL,   -- 1

    ANOMES     INT          NOT NULL,   -- 202601
    ANOMES_D   VARCHAR(10)  NOT NULL,   -- 2026-01
    DT_        VARCHAR(12)  NOT NULL,   -- 20260101

    INST       DECIMAL(18,6) NOT NULL,  -- 0,3 (convertido para 0.3)
    VB         DECIMAL(18,6) NOT NULL,  -- 0,3 (convertido para 0.3)

    MOVEL      DECIMAL(18,6) NULL,      -- existe no arquivo, então incluído

    DD_WEEK    INT NOT NULL,            -- 4
    WEEK_MM    INT NOT NULL,            -- 1
    WEEK_TXT   VARCHAR(10) NOT NULL,    -- W1
    REFERENCIA VARCHAR(45) NOT NULL     -- BRASIL
)
ENGINE=InnoDB
DEFAULT CHARSET = utf8mb4;
```

# ADICIONAR AO BANCO O CSV

```python
LOAD DATA LOCAL INFILE 'C:\\Users\\F290071\\Downloads\\2025_2026 1.csv'
INTO TABLE db_projetos.pdu
CHARACTER SET utf8mb4
FIELDS TERMINATED BY ';' ENCLOSED BY '"' ESCAPED BY '\\'
LINES TERMINATED BY '\n'
IGNORE 1 LINES
(@DT, @AAAA, @MMM, @MM, @DD, @ANOMES, @ANOMES_D, @DT_, @INST, @VB, @MOVEL, @DD_WEEK, @WEEK_MM, @WEEK_TXT, @REFERENCIA)
SET
  DT         = @DT,
  AAAA       = @AAAA,
  MMM        = @MMM,
  MM         = @MM,
  DD         = @DD,
  ANOMES     = @ANOMES,
  ANOMES_D   = @ANOMES_D,
  DT_        = @DT_,
  INST       = REPLACE(@INST, ',', '.'),
  VB         = REPLACE(@VB, ',', '.'),
  MOVEL      = NULLIF(REPLACE(@MOVEL, ',', '.'), ''),
  DD_WEEK    = @DD_WEEK,
  WEEK_MM    = @WEEK_MM,
  WEEK_TXT   = @WEEK_TXT,
  REFERENCIA = @REFERENCIA;pdu
```
